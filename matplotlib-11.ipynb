```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
m1_student_age = []
m2_student_age = []

m1_student_age = np.random.randint(18, 45, (100))
m2_student_age = np.random.randint(15, 440, (100))

print(m1_student_age)
print(m2_student_age)
```

```python
plt.hist(m1_student_age)

plt.title("ML Students Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No Of Student")
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(m1_student_age, bins, rwidth=0.8, histtype="bar", color = 'y',label="ML Student")
plt.hist(m2_student_age, bins, rwidth=0.8, histtype="bar", color = 'r',label="Python Student")
plt.title("ML Students Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No Of Student")
plt.legend()
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
# plt.figure(figsize = (16,9))
plt.hist([m1_student_age, m2_student_age], 
         bins, rwidth=0.8, histtype="bar", color = ['m', 'y'],
         label=["ML Student", "Python Student"])

plt.title("ML Students Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No Of Student")
plt.legend(loc='upper left')
plt.show()
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class_1 = [30, 10, 20, 25, 10]
class_2 = [40, 5, 30, 20, 60]
class_3 = [35, 5, 35, 15, 15]
```

```python
plt.bar(classes, class_1)
```

```python
plt.bar(classes, class_1, width = 0.2, align = "edge", color = "y", linewidth = 5)
```

```python
plt.bar(classes, class_1, width = 0.2, color = "b",
        label =" Class 1 Students") #visible=False
 
plt.bar(classes, class_2, width = 0.2, color = "g",
        label =" Class 2 Students") 
 
plt.title("Bar Chart of Class")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.show()
```

```python
classes_index = np.arange(len(classes))
 
width = 0.2
 
plt.bar(classes_index, class_1, width , color = "b",
        label =" Class 1") #visible=False
 
plt.bar(classes_index + width, class_2, width , color = "g",
        label =" Class 2") 
 
plt.bar(classes_index + width + width, class_3, width , color = "y",
        label =" Class 2") 
 
plt.xticks(classes_index + width, classes, rotation = 20)
plt.title("Bar Chart of Class")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

```python
df_google_play_store_apps = pd.read_csv("googleplaystore.csv", nrows = 1000)
df_google_play_store_apps.shape ## Get shape of dataset in rows and colms
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
plt.scatter(x,y) # Draw scatter plot with parameter x and y
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
           edgecolors="g" )#verts="<"
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
           edgecolors="g" )#verts="<"
 
plt.scatter(x,df_google_play_store_apps["Installs"], c = "y", marker = "o", s = 100, alpha=0.5, linewidths=10,
           edgecolors="c" )
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.show()
```
